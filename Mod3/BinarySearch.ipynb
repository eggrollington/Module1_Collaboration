```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0 
        high = len(arr) -1

        while low <= high:
            mid = (low + high) // 2
            if arr[mid] == k:
                while arr[mid] == k:
                    mid -=1
                    
                mid +=1
                return mid
            elif arr[mid] < k:
                low = mid +1
            else:
                high = mid - 1
        return -1
```
