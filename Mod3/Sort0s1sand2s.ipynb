```python
class Solution:
    def sort012(self, arr):
        # code here
        zeros = []
        ones = []
        twos = []
        for number in arr:
            if number == 0:
                zeros.append(number)
            elif number == 1:
                ones.append(number)
            else:
                twos.append(number)
        arr[:] = zeros + ones + twos
        return arr
```
